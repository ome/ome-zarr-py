```python
import os
import shutil
from skimage.filters import threshold_otsu
import dask.array as da
from dask import delayed

from ome_zarr.image import NgffImage, NgffMultiscales
from ome_zarr.writer import write_image
```

```python
dummy_array = da.random.random(
    size=(10, 3, 128, 128, 128), chunks=(1, 3, 64, 64, 64)
)

dummy_labels = dummy_array > 0.5
```

```python
ngff_image = NgffImage(
    data=dummy_array,
    dims=["t", "c", "z", "y", "x"],
    scale={"x": 0.1, "y": 0.1, "z": 0.5},
)

ngff_multiscales = NgffMultiscales(
    image=ngff_image,
    scale_factors=[2, 4, 8],
)

ngff_labels = NgffImage(
    data=dummy_labels,
    dims=["t", "c", "z", "y", "x"],
    scale={"x": 0.1, "y": 0.1, "z": 0.5},
    name="some_segmentation_labels",
)
ngff_multiscales_labels = NgffMultiscales(
    image=ngff_labels,
    scale_factors=[2, 4, 8],
    method="nearest",
)
```

```python
ngff_multiscales.to_ome_zarr(
  "test.ome.zarr",
  labels=ngff_multiscales_labels,)
```
