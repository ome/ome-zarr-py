```python
import dask.array as da
from ome_zarr.writer import write_image
from ome_zarr.image import NgffImage, NgffMultiscales
import shutil
import os
```

```python
dummy_array = da.random.random(
    size=(10, 3, 128, 128, 128), chunks=(1, 3, 64, 64, 64)
)
```

```python
if os.path.exists("test.ome.zarr"):
    shutil.rmtree("test.ome.zarr")

write_image(
    image=dummy_array,
    group="test.ome.zarr",
    axes="tczyx",
    scale={"x": 0.1, "y": 0.1, "z": 0.5},
)
```

```python
ngff_image = NgffImage(
    data=dummy_array,
    dims=["t", "c", "z", "y", "x"],
    scale={"x": 0.1, "y": 0.1, "z": 0.5},
)

ngff_multiscales = NgffMultiscales(
    image=ngff_image,
    scale_factors=[2, 4, 8],
)

ngff_multiscales
```

```python
ngff_multiscales.to_ome_zarr("test.ome.zarr")
```
